```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import plotly.express as px
from matplotlib.lines import Line2D  
import numpy as np
```

```python
df_n1 = pd.read_csv("nexus1.csv", delimiter=',', skiprows=0, low_memory=False)
print(df_n1.shape)
print(df_n1)
```

```python
df_n2 = pd.read_csv("nexus2.csv", delimiter=',', skiprows=0, low_memory=False)
print(df_n2.shape)
print(df_n2.columns)
```

```python
def plots_nexus(df):
    # Convert DataFrame columns to NumPy arrays
    time = df['%time'].values
    ux = df['field.ux'].values
    uy = df['field.uy'].values
    px = df['field.px'].values
    py = df['field.py'].values

    # Plot ux vs time
    plt.figure(figsize=(10, 5))
    plt.plot(time, ux, marker='o', linestyle='-')
    plt.title('ux vs Time')
    plt.xlabel('Time')
    plt.ylabel('ux')
    plt.grid(True)
    plt.axis('equal')
    plt.show()

    # Plot uy vs time
    plt.figure(figsize=(10, 5))
    plt.plot(time, uy, marker='o', linestyle='-')
    plt.title('uy vs Time')
    plt.xlabel('Time')
    plt.ylabel('uy')
    plt.axis('equal')
    plt.grid(True)
    plt.show()

    # Plot position (px, py)
    plt.figure(figsize=(10, 5))
    plt.plot(px, py, marker='o', linestyle='-')
    plt.title('Position (px, py)')
    plt.xlabel('px')
    plt.ylabel('py')
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
plots_nexus(df_n1)
```

```python
plots_nexus(df_n2)
```

```python

px1 = df_n1['field.px'].values
py1 = df_n1['field.py'].values

px2 = df_n2['field.px'].values
py2 = df_n2['field.py'].values

# Plot position (px, py)
plt.figure(figsize=(10, 5))
plt.plot(px1, py1, marker='o', linestyle='-',  label='nexus1')
plt.plot(px2, py2, marker='o', linestyle='-', label='nexus2')
plt.title('Position (px, py)')
plt.xlabel('px')
plt.ylabel('py')
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def plot_vectors(df):
    px = df['field.px'].values
    py = df['field.py'].values
    ux = df['field.ux'].values
    uy = df['field.uy'].values

    plt.figure(figsize=(10, 7))
    plt.title('Vector Plot')
    plt.xlabel('px')
    plt.ylabel('py')

    # Plot position points
    # plt.scatter(px, py, color='red', label='Position')

    # Calculate the magnitude (Euclidean norm) of the vectors
    vector_magnitude = np.sqrt(np.square(ux) + np.square(uy))

    # Plot vectors
    plt.quiver(px, py, ux, uy, vector_magnitude, angles='xy', scale_units='xy', scale=1, cmap='viridis', label='Control Vector')

    plt.grid(True)
    plt.colorbar(label='Magnitude of Velocity Vector')
    plt.legend()
    plt.show()
```

```python
plot_vectors(df_n1)
plot_vectors(df_n2)
```
